```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import datetime
import time
import random
import json
import pickle
```

```python
cd ../backup/mongodb_backup/
```

```python
with open("sales_order_details_08-01-18.json") as file:
    f = file.readlines()
```

```python
json_data = [json.loads(i) for i in f]
```

```python
for i, data in enumerate(json_data):
    json_data[i]['Sales Order Date'] = datetime.datetime.fromtimestamp(int(json_data[i]['Sales Order Date']['$date']['$numberLong'][:10]))
    json_data[i]['Sales Order Expiry Date'] = datetime.datetime.fromtimestamp(int(json_data[i]['Sales Order Expiry Date']['$date']['$numberLong'][:10]))
```

```python
df = pd.DataFrame.from_dict(json_data)
```

```python
df.drop(columns = ['_id', 'Sales Order No'], inplace = True)
df['Sales Order Quantity'] = pd.to_numeric(df['Sales Order Quantity'])
```

### Total Sales Order Quantity wrt. Products and Scheme

```python
df['Product & Scheme'] = df['Grade&Size(Product Description)'] + '|' + df['Scheme(FSA/Auction etc)']
df_temp = df[['Product & Scheme', 'Sales Order Quantity']].groupby(df['Product & Scheme']).sum().sort_values('Sales Order Quantity', ascending = False).reset_index()
df_temp['Product'] = df_temp['Product & Scheme'].apply(lambda x : x.split('|')[0])
df_temp['Scheme'] = df_temp['Product & Scheme'].apply(lambda x : x.split('|')[1])
df_temp.sort_values('Product')[['Product', 'Scheme', 'Sales Order Quantity']]
```

### Total No. Of Consignments (wrt to Product and Scheme)

```python
df_temp = pd.crosstab(df['Grade&Size(Product Description)'], df['Scheme(FSA/Auction etc)'])
df_temp['Total'] = df_temp['COKERY'] + df_temp['Eauction'] + df_temp['MOU']
df_temp.sort_values('Total', ascending = False)[:20]
```
