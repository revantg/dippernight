```python
import numpy as np
import pandas as pd
import scipy 
import matplotlib.pyplot as plt
import seaborn as sns
import random
import json
import datetime, time
import pickle
```

```python
with open("daily_dispatch_summary_18-01-19.json") as f:
    a = f.readlines()
    json_data = [json.loads(i) for i in a]
```

```python
for i in range(len(json_data)):
    json_data[i]['Sales Order Date'] = json_data[i]['Sales Order Date']['$date']['$numberLong']
    json_data[i]['Loading Date'] = json_data[i]['Loading Date']['$date']['$numberLong']
```

```python
df = pd.DataFrame.from_dict(json_data).drop(columns = ['_id', 'HEAP ID'])
```

```python
df['Sales Order Date'] = df['Sales Order Date'].apply(lambda x : datetime.datetime.fromtimestamp(int(x[:10])))
df['Loading Date'] = df['Loading Date'].apply(lambda x : datetime.datetime.fromtimestamp(int(x[:10])))

df['Quality Lifted'] = df['Quality Lifted'].astype('float')
```

```python
delay = [i- j for i, j in df[['Loading Date', 'Sales Order Date']].values]
df['delay'] = delay
df['delay'] = df['delay'].astype('timedelta64[D]')
```

```python
df['delay-bracket'] = df['delay'].apply(lambda x : str((x//20) * 20) + ' - ' + str(((x//20) * 20) + 20))
```
